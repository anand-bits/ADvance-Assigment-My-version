```python
# Dividing The Data in Numerical and Object Feature

numeric_feature=[feature for feature in df.columns if df[feature].dtype!= object]
numeric_feature
```

```python
##Univariate Analysis

plt.figure(figsize=(15,15))
plt.suptitle("Univariate Analysis Of Numerical Features",fontsize=20)

for i in range (0,len(numeric_feature)):
  plt.subplot(5,3,i+1)
  sns.kdeplot(x=df[numeric_feature[i]],shade=True,color='r')
  plt.xlabel(numeric_feature[i])
  plt.tight_layout()
```
